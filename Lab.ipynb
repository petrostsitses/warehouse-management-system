```python
import csv
import json

                     # =====Σύστημα Διαχείρησης Αποθήκης====

inventory= []

def add_product(product_id,name,price,quantity,category):
    product={"id":product_id, "name":name, "price":price, "quantity":quantity, "category":category}
    inventory.append(product)
    print(f"Το προιον '{name}' προστέθηκε με επιτυχία",)

add_product(1,"Ποντίκι Logitech M185", 15.99, 40, "Περιφερειακά"),
add_product(2,"Πληκτρολόγιο Μηχανικό RedDragon", 49.90, 25, "Περιφερειακά"),
add_product(3,"Οθόνη Dell 24''", 139.99, 10, "Οθόνες" ),
add_product(4, "Laptop Lenovo IdeaPad 3", 599.00, 8, "Υπολογιστές"),
add_product(5, "USB Stick 64GB Kingston", 9.99, 50, "Αποθηκευτικά Μέσα"),
add_product(6, "Καλώδιο HDMI 2m" ,5.50, 30, "Αξεσουάρ"),
add_product(7, "Router TP-Link",  39.90, 20, "Δίκτυο"),
add_product(8, "Κάρτα μνήμης SD 128GB", 12.90, 20, "Αποθηκευτικά Μέσα"),
add_product(9, "MousePad", 4.00, 70, "Περιφερειακά"),
add_product(10, "Μπαταρίες AA (4τμχ)", 3.50, 60, "Αξεσουάρ")



                            #===Εμφάνιση Αποθήκης===
def display_inventory():
     if len(inventory) == 0:
        print("Η Αποθήκη είναι άδεια")

     else:
        print("\nΠεριεχόμενα Αποθήκης")
        for product in inventory:
         print(f"ID:{product['id']}, Όνομα:{product['name']}, Τιμή:{product['price']}, Απόθεμα:{product['quantity']}, Κατηγορία:{product['category']} ")

display_inventory()


                        #===Διαγραφή Προιόντων===

def remove_product(product_id):
    for product in inventory:
        if product['id'] == product_id:
             inventory.remove(product)
             print(f"Το Προιον με ID {product_id} αφαιρέθηκε απο την αποθήκη.")
             return
    print(f"Δεν βρέθηκε προιον με ID{product_id}.")


                           #===Αναζήτηση Προιόντων===

def search_product(product_name):
    results = [p for p in inventory if p['name'].lower() == product_name.lower()]
    if results:
            print("Βρέθηκαν τα προιοντα")
            for p in results:
                print(f"ID:{p['id']}, Όνομα:{p['name']}, Τιμή:{p['price']}€, Απόθεμα:{p['quantity']}")
    else:
          print("Το προιον δεν βρέθηκε")



                           #===Ενημέρωση Προιόντων===

def update_product(product_id):
    for product in inventory:
        if product['id'] == product_id:
            new_price = input(f"Νέα τιμή για '{product['name']}' : ")
            if new_price.strip() != "":
                product['price'] = float(new_price)

            new_quantity = input(f"Νέα ποσότητα για '{product['name']}' : ")
            if new_quantity.strip() != "":
                product['quantity'] = int(new_quantity)

            print(f"Το προϊόν με ID {product_id} ενημερώθηκε")
            return

    print(f"Δεν βρέθηκε προϊόν με ID {product_id}.")


                             #=== Υπολογισμός Αξίας Αποθήκης===

def calculate_total_value():
    total = 0
    for product in inventory:
        total= total + product['price'] * product['quantity']
    print(f"Η Συνολική αξία της αποθήκης είναι: {total:.2f}€")



                        #==== Αποθήκευση Δεδομένων====


def save_to_file(filename="inventory.json"):
    with open(filename, "w", encoding="utf-8") as f:
        json.dump(inventory, f, ensure_ascii=False, indent=4)
    print(f"Τα δεδομένα αποθηκεύτηκαν στο αρχείο {filename}")




                     #==== Φόρτωση Δεδομένων ====

def load_from_file(filename="inventory.json"):
    global inventory
    try:
        with open(filename, "r",encoding="utf-8") as f:
            inventory = json.load(f)
        print(f"Τα δεδομένα φορτώθηκαν από το αρχείο {filename}")
    except FileNotFoundError:
        print("Δεν βρέθηκε αρχείο. Η αποθήκη φαίνεται άδεια.")



              #==== Εξαγωγή και Εισαγωγή σε CSV ====
def export_csv(filename="inventory.csv"):
    with open(filename, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow(["id", "name", "price", "quantity", "category"])
        for product in inventory:
            writer.writerow([product['id'], product['name'], product['price'], product['quantity'], product['category']])
    print(f"Τα δεδομένα εξήχθησαν στο αρχείο {filename}")




import csv

def import_from_csv(filename="inventory.csv"):
    try:
        with open(filename, "r", encoding="utf-8-sig") as f:
            reader = csv.DictReader(f)
            for row in reader:
                # Έλεγχος για διπλό ID
                if not any(p['id'] == int(row['id']) for p in inventory):
                    product = {
                        "id": int(row['id']),
                        "name": row['name'],
                        "price": float(row['price']),
                        "quantity": int(row['quantity']),
                        "category": row['category']
                    }
                    inventory.append(product)
        print(f"Τα δεδομένα φορτώθηκαν από το αρχείο {filename} με επιτυχία.")
    except FileNotFoundError:
        print(f"Το αρχείο {filename} δεν βρέθηκε.")
    except Exception as e:
        print(f"Προέκυψε σφάλμα κατά την εισαγωγή: {e}")






                         #==== Ταξινόμηση Προιόντων ====

def sort_inventory(by="name"):
    if by == "name":
        sorted_inventory = sorted(inventory, key=lambda k: k['name'].lower())
    elif by == "price":
        sorted_inventory = sorted(inventory, key=lambda k: k['price'])

    elif by == "quantity":
        sorted_inventory = sorted(inventory, key=lambda k: k['quantity'])

    elif by == "value":
        sorted_inventory = sorted(inventory, key=lambda k: k['price'] * k['quantity'])

    else:
        sorted_inventory = inventory

    if not sorted_inventory:
        print("Η αποθήκη είναι άδεια")
    else:
        print("\nΠεριεχόμενα Αποθήκης (Ταξινομημένα)")
        for product in sorted_inventory:
            print(f"ID:{product['id']}, Όνομα:{product['name']}, Τιμή:{product['price']}, Απόθεμα:{product['quantity']}, Κατηγορία:{product['category']}")


                         #==== Κατηγοριοποίηση Προιόντων ====

def categorize_product(by="name"):
    if by == "name":
        categorized_product = sorted(inventory, key=lambda k: k['name'].lower())
    elif by == "price":
        categorized_product = sorted(inventory, key=lambda k: k['price'])
    else:
        categorized_product = inventory.copy()

    if not categorized_product:
        print("Δεν υπάρχουν αποτελέσματα!")

    else:
        print("\nΚατηγοριοποίηση Προιόντων")
        for p in categorized_product:
            print(f"ID:{p['id']}, Όνομα:{p['name']}, Τιμή:{p['price']}, Απόθεμα:{p['quantity']}, Κατηγορία:{p['category']}")







                            #==== Φιλτράρισμα Προιόντων ====

def filter_inventory(threshold):
    filtered_inventory = []

    for product in inventory:
        if product['quantity'] < threshold:
            filtered_inventory.append(product)

    if not filtered_inventory:
        print("Δεν βρέθηκαν προϊόντα με χαμηλό απόθεμα.")
    else:
        print("\nΠροϊόντα με χαμηλό απόθεμα:")
        for product in filtered_inventory:
            print(f"ID:{product['id']}, Όνομα:{product['name']}, Τιμή:{product['price']}, Απόθεμα:{product['quantity']}, Κατηγορία:{product['category']}")




                      #==== Ανάλυση Αποθέματος ====
def analyze_inventory(inventory):
    if len(inventory) == 0:
        print("Η Αποθήκη είναι άδεια")
        return

    most_expensive_product = max(inventory, key=lambda k: k['price'])
    most_stocked_product = max(inventory, key=lambda k: k['quantity'])
    totals = category_totals(inventory)  # παίρνει το λεξικό {κατηγορία: συνολική αξία}
    max_category = max(totals, key=totals.get)

    print(f"Συνολικά προϊόντα: {len(inventory)}")
    print("Αξία ανά κατηγορία:")
    for cat, val in totals.items():
        print(f"{cat}: {val:.2f}€")
    print(f"Πιο ακριβό προϊόν: {most_expensive_product['name']} - Τιμή: {most_expensive_product['price']}")
    print(f"Προϊόν με μεγαλύτερη ποσότητα: {most_stocked_product['name']} - Απόθεμα: {most_stocked_product['quantity']}")
    print(f"Κατηγορία με μεγαλύτερη συνολική αξία: {max_category} - Αξία: {totals[max_category]}")


def category_totals(inventory):
    category_totals = {}

    for product in inventory:
        value = product['price'] * product['quantity']
        cat = product['category']

        if cat in category_totals:
            category_totals[cat] += value
        else:
            category_totals[cat] = value

    return category_totals



            #==== Γραφήματα ====

         #=== Bar chart ===

import matplotlib.pyplot as plt

def plot_quantity_per_product():
    names = [p['name'] for p in inventory]
    quantities = [p['quantity'] for p in inventory]

    plt.figure(figsize=(10,6))
    plt.bar(names, quantities, color='skyblue')
    plt.xlabel("Προϊόν")
    plt.ylabel("Ποσότητα")
    plt.title("Ποσότητα Αποθέματος ανά Προϊόν")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()


        #=== Pie chart ===

def plot_value_per_category():
    totals = category_totals(inventory)  # παίρνουμε το λεξικό {κατηγορία: συνολική αξία}
    categories = list(totals.keys())
    values = list(totals.values())

    plt.figure(figsize=(8,8))
    plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140)
    plt.title("Κατανομή Αξίας Αποθέματος ανά Κατηγορία")
    plt.legend(categories, title="Κατηγορίες", loc="best")
    plt.tight_layout()
    plt.show()


             #==== Scatter Plot ====
def plot_price_vs_quantity():
    categories = list(set(p['category'] for p in inventory))
    colors = plt.cm.tab10.colors
    sizes = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500]

    # Συνδέουμε κατηγορία -> μέγεθος
    category_size = {cat: sizes[i % len(sizes)] for i, cat in enumerate(categories)}

    category_color = {cat: colors[i % len(colors)] for i, cat in enumerate(categories)}

    plt.figure(figsize=(8,6))
    for p in inventory:
        plt.scatter(p['price'], p['quantity'], color=category_color[p['category']],
                    s=category_size[p['category']], alpha=0.7, label=p['category'])

    # για να μην επαναλαμβάνεται η legend
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), title="Κατηγορίες")

    plt.xlabel("Τιμή")
    plt.ylabel("Ποσότητα")
    plt.title("Σχέση Τιμής – Ποσότητας Προϊόντων")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()



                #===Κυρίως Πρόγραμμα Main menu===
def main():
     while True:
         print("\n====Σύστημα Διαχείρησης Αποθήκης====")
         print("1. Προσθήκη Προιόντος")
         print("2. Διαγραφή Προιόντος")
         print("3. Αναζήτηση Προιόντος")
         print("4. Ενημέρωση Προιόντος")
         print("5. Υπολογισμός Αξίας Αποθήκης")
         print("6. Εμφάνιση Αποθήκης")
         print("7. Αποθήκευση Δεδομένων")
         print("8. Φόρτωση Δεδομένων")
         print("9. Ταξινόμηση κατα όνομα")
         print("10. Ταξινόμηση κατα Τιμή")
         print("11. Ταξινόμηση κατά Ποσότητα")
         print("12. Ταξινόμηση κατά Αξία")
         print("13. Κατηγοριοποίηση κατα Όνομα")
         print("14. Κατηγοριοποίηση κατα Τιμή")
         print("15. Φίλτρο Χαμηλού αποθέματος")
         print("16. Ανάλυση Αποθέματος")
         print("17. Εξαγωγή αποθήκης σε CSV")
         print("18. Εισαγωγή αποθήκης σε CSV")
         print("19. Bar chart: Ποσότητα ανά προϊόν")
         print("20. Pie chart: Αξία ανά κατηγορία")
         print("21. Scatter Plot Τιμής-Ποσότητας")
         print("0. Έξοδος")

         choice = input ("Επιλέξτε μια ενέργεια: ")

         if choice == "1":
             try:
                 product_id=int(input("Δώστε id Προιόντος:  "))
                 name= input("Δώστε όνομα προιόντος: ")
                 price= float(input("Δώστε τιμή προιόντος: "))
                 quantity= int(input("Δώστε απόθεμα προιόντος: "))
                 category = input("Δώστε κατηγορία προιόντος: ")
                 add_product(product_id,name,price,quantity,category)
             except ValueError:
                print("\nΛάθος είσοδος! Βεβαιωθείτε ότι οι τιμές είναι σωστού τύπου.")

         elif choice == "2":
             try:
                 product_id=int(input("\nΔώστε το ID του προϊόντος που θέλετε να διαγράψετε:  "))
                 remove_product(product_id)
             except ValueError:
                 print("\nΛάθος! Το ID πρέπει να είναι αριθμός.")

         elif choice == "3":
             try:
                 product_name=input("\nΑναζητήστε το προιον με το όνομα:  ")
                 search_product(product_name)
             except ValueError:
                 print("\nΛάθος! Παρακαλώ δώσε το όνομα του προιόντος για αναζήτηση")

         elif choice == "4":
             try:
                 product_id=int(input("\nΔώστε ID για να ενημερώσετε το προιον: "))
                 update_product(product_id)
             except ValueError:
                 print("\nΛάθος! Το ID πρέπει να είναι αριθμός.")

         elif choice == "5":
             print("\nΗ Συνολική αξία της αποθήκης είναι: ")
             calculate_total_value()

         elif choice == "6":
             print("\nΠεριεχόμενα Αποθήκης: ")
             display_inventory()


         elif choice == "7":
             print("\nΑποθήκευση Δεδομένων:  ")
             save_to_file(filename="inventory.json")


         elif choice == "8":
             print("\nΦόρτωση Δεδομένων: ")
             load_from_file(filename="inventory.json")


         elif choice == "9":
             print("\nΤαξινόμηση κατα Όνομα: ")
             sort_inventory(by="name")

         elif choice == "10":
             print("\nΤαξινόμηση κατα Τιμή: ")
             sort_inventory(by="price")

         elif choice == "11":
             print("\nΤαξινόμηση κατα Ποσότητα: ")
             sort_inventory(by="quantity")


         elif choice == "12":
             print("\nΤαξινόμηση κατά Αξία: ")
             sort_inventory(by="value")


         elif choice == "13":
             print("\nΚατηγοριοποίηση Προιόντων κατα Όνομα: ")
             categorize_product(by="name")

         elif choice == "14":
             print("\nΚατηγοριοποίηση κατα Τιμή: ")
             categorize_product(by="price")

         elif choice == "15":
          try:
            threshold = int(input("Δώστε όριο για χαμηλό απόθεμα: "))
            filter_inventory(threshold)
          except ValueError:
             print("Λάθος! Το όριο πρέπει να είναι αριθμός.")


         elif choice == "16":
           print("\nΑνάλυση Αποθέματος")
           analyze_inventory(inventory)


         elif choice == "17":
           print("\nΕξαγωγή αποθήκης σε CSV")
           export_csv()

         elif choice == "18":
           print("\nΕισαγωγή αποθήκης από CSV")
           import_from_csv()

         elif choice == "19":
             print("\nBar chart: Ποσότητα ανά προϊόν")
             plot_quantity_per_product()

         elif choice == "20":
             print("\nPie chart: Αξία ανά κατηγορία")
             plot_value_per_category()

         elif choice == "21":
             print("\nScatter plot: Τιμή – Ποσότητα ανά κατηγορία")
             plot_price_vs_quantity()

         elif choice == "0":
             print("\nΈξοδος απο το πρόγραμμα")
             break
         else:
             print("\nΜη έγκυρη επιλογή! Δοκιμάστε ξανά")


if __name__ == "__main__":
    main()









```
